```python
# final code for All applications.....................................................................
import tkinter as tk
from tkinter import messagebox, scrolledtext
import time
import pyttsx3
import speech_recognition as sr
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QToolBar, QAction, QLineEdit
from PyQt5.QtCore import QUrl
from PyQt5.QtWebEngineWidgets import QWebEngineView
from PyQt5.QtGui import QKeySequence
from PyQt5.QtCore import Qt
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog
import os
from PIL import Image, ImageTk
import tkinter as tk
from datetime import datetime
from tkcalendar import Calendar
import pyttsx3
import tkinter as Tkinter
from datetime import datetime
import pyttsx3
import tkinter as tk
from tkinter import scrolledtext
import speech_recognition as sr
import pyttsx3
import tkinter as tk
from tkinter import messagebox
import time
import pyttsx3
import tkinter as tk
import speech_recognition as sr
import pyttsx3
import tkinter as Tkinter
from datetime import datetime
import pyttsx3
import tkinter as tk
from tkinter import scrolledtext
import speech_recognition as sr
import pyttsx3
import cv2
import os
import pyttsx3
import pyaudio
import wave
import threading
from moviepy.editor import VideoFileClip, AudioFileClip
from tkinter import *
from PIL import Image, ImageTk
import os
import speech_recognition as sr
import pyttsx3
import keyboard
import subprocess
import platform
import tkinter as tk
from tkinter import ttk, simpledialog
from tkinter import messagebox
import time
import pyttsx3


def run_calculator():
    class Calculator:
        def __init__(self, master):
            self.master = master
            self.master.title('Calculator')
            self.equation = ""
            self.speech_engine = pyttsx3.init()
            self.recognizer = sr.Recognizer()
            self.stop_listening = None
            self.is_listening = False

            # Entry field to show calculations
            self.entry_field = tk.Entry(master, width=40, borderwidth=5, font=('arial', 20, 'bold'))
            self.entry_field.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky='we')

            # Buttons
            buttons = [
                '7', '8', '9', '/',
                '4', '5', '6', '*',
                '1', '2', '3', '-',
                '0', '.', '=', '+'
            ]

            row_val = 1
            col_val = 0

            for button in buttons:
                tk.Button(master, text=button, width=10, font=('arial', 20, 'bold'), command=lambda button=button: self.click_button(button)).grid(row=row_val, column=col_val, sticky='we')
                col_val += 1
                if col_val > 3:
                    col_val = 0
                    row_val += 1

            # Clear and backspace buttons
            clear_button = tk.Button(master, text='Clear', width=10, font=('arial', 20, 'bold'), command=self.clear_all)
            clear_button.grid(row=row_val, column=0, columnspan=2, sticky='we')

            backspace_button = tk.Button(master, text='Backspace', width=10, font=('arial', 20, 'bold'), command=self.backspace)
            backspace_button.grid(row=row_val, column=2, columnspan=2, sticky='we')

            # Bind keys to start and stop listening
            master.bind('<Control-Shift-KeyPress-V>', self.start_listening)
            master.bind('<Control-Shift-KeyRelease-V>', self.stop_listening_and_calculate)
            master.bind('<BackSpace>', self.backspace_input)
            master.bind('<Delete>', self.clear_all_input)
            master.bind('<Key>', self.keyboard_input)
            master.bind('<Return>', self.calculate_result)

        def click_button(self, button):
            if button == '=':
                # Perform calculation
                try:
                    self.equation = str(eval(self.equation))
                    self.speech_engine.say("The result is " + self.equation)
                    self.speech_engine.runAndWait()
                except:
                    self.equation = "Error"
                    self.speech_engine.say("Error")
                    self.speech_engine.runAndWait()
            else:
                # Add button press to equation
                self.equation += str(button)
                self.speech_engine.say(str(button))
                self.speech_engine.runAndWait()
            self.update_entry_field()

        def clear_all(self):
            self.equation = ""
            self.update_entry_field()

        def backspace(self):
            self.equation = self.equation[:-1]
            self.update_entry_field()

        def start_listening(self, event):
            if not self.is_listening:
                self.is_listening = True
                with sr.Microphone() as source:
                    audio = self.recognizer.listen(source)
                    try:
                        command = self.recognizer.recognize_google(audio)
                        self.equation += command
                        self.update_entry_field()
                    except sr.UnknownValueError:
                        pass
                    except sr.RequestError:
                        pass

        def stop_listening_and_calculate(self, event):
            if self.is_listening:
                self.is_listening = False
                self.click_button('=')

        def clear_all_input(self, event):
            self.clear_all()

        def backspace_input(self, event):
            self.backspace()

        def update_entry_field(self):
            self.entry_field.delete(0, tk.END)
            self.entry_field.insert(0, self.equation)
            self.speech_engine.say(self.equation)
            self.speech_engine.runAndWait()

        def keyboard_input(self, event):
            key = event.keysym
            if key.isdigit() or key in ['+', '-', '*', '/', '.', '=', 'BackSpace', 'Delete']:
                if key == 'BackSpace':
                    self.backspace()
                elif key == 'Delete':
                    self.clear_all()
                elif key == '=':
                    self.click_button('=')
                else:
                    self.click_button(key)
            elif key == 'plus':
                self.click_button('+')
            elif key == 'minus':
                self.click_button('-')
            elif key == 'asterisk':
                self.click_button('*')
            elif key == 'slash':
                self.click_button('/')
            else:
                pass

        def calculate_result(self, event):
            self.click_button('=')

    root = tk.Tk()
    root.state('zoomed')  # This line makes the window maximized
    root.grid_columnconfigure(0, weight=1)  # These lines make the widgets expand when the window size changes
    root.grid_columnconfigure(1, weight=1)
    root.grid_columnconfigure(2, weight=1)
    root.grid_columnconfigure(3, weight=1)
    root.grid_rowconfigure(0, weight=1)
    root.grid_rowconfigure(1, weight=1)
    root.grid_rowconfigure(2, weight=1)
    root.grid_rowconfigure(3, weight=1)
    root.grid_rowconfigure(4, weight=1)
    root.grid_rowconfigure(5, weight=1)
    my_calculator = Calculator(root)
    root.mainloop()



def run_web_browser():
    class WebBrowser(QMainWindow):
        def __init__(self):
            super().__init__()

            # Initialize the web browser view
            self.browser = QWebEngineView()
            self.browser.setUrl(QUrl("http://www.google.com"))

            # Set the web browser as the central widget
            self.setCentralWidget(self.browser)

            # Create navigation bar
            navbar = QToolBar()
            self.addToolBar(navbar)

            # Back Button
            back_btn = QAction('Back', self)
            back_btn.setStatusTip('Back to previous page')
            back_btn.triggered.connect(self.browser.back)
            navbar.addAction(back_btn)

            # Forward Button
            forward_btn = QAction('Forward', self)
            forward_btn.setStatusTip('Forward to next page')
            forward_btn.triggered.connect(self.browser.forward)
            navbar.addAction(forward_btn)

            # Reload Button
            reload_btn = QAction('Reload', self)
            reload_btn.setStatusTip('Reload page')
            reload_btn.triggered.connect(self.browser.reload)
            navbar.addAction(reload_btn)

            # URL Bar
            self.url_bar = QLineEdit()
            self.url_bar.returnPressed.connect(self.navigate_to_url)
            navbar.addWidget(self.url_bar)

            # Home Button
            home_btn = QAction('Home', self)
            home_btn.setStatusTip('Go home')
            home_btn.triggered.connect(self.navigate_home)
            navbar.addAction(home_btn)

            # Stop Button
            stop_btn = QAction('Stop', self)
            stop_btn.setStatusTip('Stop loading current page')
            stop_btn.triggered.connect(self.browser.stop)
            navbar.addAction(stop_btn)

            # URL Changed Event
            self.browser.urlChanged.connect(self.update_urlbar)

            # Go Back, Go Forward, Refresh, URL Bar Context Menu
            navbar.addSeparator()

            # Go Back Button
            back_btn = QAction('Go Back', self)
            back_btn.triggered.connect(self.browser.back)
            navbar.addAction(back_btn)

            # Go Forward Button
            forward_btn = QAction('Go Forward', self)
            forward_btn.triggered.connect(self.browser.forward)
            navbar.addAction(forward_btn)

            # Refresh Button
            refresh_btn = QAction('Refresh', self)
            refresh_btn.triggered.connect(self.browser.reload)
            navbar.addAction(refresh_btn)

            # Clear URL Bar Button
            clear_url_btn = QAction('Clear URL', self)
            clear_url_btn.triggered.connect(self.clear_urlbar)
            navbar.addAction(clear_url_btn)

            # Voice Command Activation Button
            voice_activation_btn = QAction('Voice Command (Space)', self)
            voice_activation_btn.setShortcut(QKeySequence(Qt.Key_Space))
            voice_activation_btn.triggered.connect(self.activate_voice_command)
            self.addAction(voice_activation_btn)

            # Initialize speech recognizer
            self.recognizer = sr.Recognizer()
            self.microphone = sr.Microphone()

        def activate_voice_command(self):
            with self.microphone as source:
                self.recognizer.adjust_for_ambient_noise(source)
                print("Listening...")
                audio = self.recognizer.listen(source)

            try:
                command = self.recognizer.recognize_google(audio).lower()
                print("Command:", command)
                self.process_voice_command(command)
            except sr.UnknownValueError:
                print("Could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))

        def process_voice_command(self, command):
            # Define commands and their corresponding actions here
            if 'go back' in command:
                self.browser.back()
            elif 'go forward' in command:
                self.browser.forward()
            elif 'refresh' in command:
                self.browser.reload()
            elif 'stop' in command:
                self.browser.stop()
            elif 'navigate' in command:
                # Extract URL from command and navigate
                url = command.split('navigate to ')[1]
                self.navigate_to_url(url)
            elif 'home' in command:
                self.navigate_home()

        def navigate_home(self):
            self.browser.setUrl(QUrl("http://www.google.com"))

        def navigate_to_url(self, url=None):
            if not url:
                url = self.url_bar.text()
            q = QUrl(url)
            if q.scheme() == "":
                q.setScheme("http")
            self.browser.setUrl(q)

        def update_urlbar(self, q):
            self.url_bar.setText(q.toString())
            self.url_bar.setCursorPosition(0)

        def clear_urlbar(self):
            self.url_bar.clear()

    app = QApplication(sys.argv)
    QApplication.setApplicationName("Python Web Browser")

    main = WebBrowser()
    main.show()

    app.exec_()

    
#final code for filemanager


def run_file_manager():
    class FileManagerApp:
        def __init__(self, master):
            self.master = master
            master.title("File Manager")

            self.header_frame = Frame(master)
            self.header_frame.pack(side=TOP, fill=X)

            self.back_button = Button(self.header_frame, text="Back", command=self.display_file_manager, width=20, height=1)
            self.back_button.pack(side=LEFT)

            self.button_frame = Frame(master)
            self.button_frame.pack(side=LEFT, fill=Y)

            self.listbox_frame = Frame(master)
            self.listbox_frame.pack(side=RIGHT, fill=BOTH, expand=True)

            self.display_file_manager()

            # Bind arrow keys to operate buttons
            master.bind("<Up>", lambda event: self.navigate_buttons(-1))
            master.bind("<Down>", lambda event: self.navigate_buttons(1))
            master.bind("<Left>", lambda event: self.navigate_files(-1))
            master.bind("<Right>", lambda event: self.navigate_files(1))
            master.bind("<Return>", lambda event: self.select_button())

            # Initialize the recognizer and the speech engine
            self.r = sr.Recognizer()
            self.engine = pyttsx3.init()

            # Add these lines to the __init__ method of FileManagerApp
            self.speaking = False
            keyboard.add_hotkey('space', self.start_listening)
            keyboard.add_hotkey('q', self.quit_app)
            keyboard.add_hotkey('ctrl + s', self.toggle_speech)

        def display_file_manager(self):
            for widget in self.button_frame.winfo_children():
                widget.destroy()

            for widget in self.listbox_frame.winfo_children():
                widget.destroy()

            picture_button = Button(self.button_frame, text="Pictures", command=lambda: self.display_file_type("Pictures"), width=20, height=1)
            video_button = Button(self.button_frame, text="Videos", command=lambda: self.display_file_type("Videos"), width=20, height=1)
            document_button = Button(self.button_frame, text="Documents", command=lambda: self.display_file_type("Documents"), width=20, height=1)
            music_button = Button(self.button_frame, text="Music", command=lambda: self.display_file_type("Music"), width=20, height=1)
            download_button = Button(self.button_frame, text="Download", command=lambda: self.display_file_type("Download"), width=20, height=1)

            self.buttons = [picture_button, video_button, document_button, music_button, download_button]
            self.selected_button_index = 0
            self.buttons[self.selected_button_index].config(highlightbackground="blue")

            for button in self.buttons:
                button.bind("<Enter>", lambda event, text=button.cget("text"): self.speak(text))
                button.pack(fill=Y, expand=True)

        def navigate_buttons(self, direction):
            self.buttons[self.selected_button_index].config(highlightbackground="SystemButtonFace")
            self.selected_button_index = (self.selected_button_index + direction) % len(self.buttons)
            self.buttons[self.selected_button_index].config(highlightbackground="blue")
            self.speak(self.buttons[self.selected_button_index].cget("text"))

        def navigate_files(self, direction):
            if hasattr(self, 'file_list'):
                current_selection = self.file_list.curselection()
                if current_selection:
                    self.file_list.selection_clear(current_selection)
                    new_index = (current_selection[0] + direction) % self.file_list.size()
                    self.file_list.selection_set(new_index)
                    self.file_list.see(new_index)
                    self.speak(self.file_list.get(new_index))

        def select_button(self):
            self.buttons[self.selected_button_index].invoke()

        def display_file_type(self, folder_name):
            for widget in self.listbox_frame.winfo_children():
                widget.destroy()

            self.file_list = Listbox(self.listbox_frame)

            home_dir = os.path.expanduser("~")
            folder_path = os.path.join(home_dir, folder_name)

            for file in os.listdir(folder_path):
                self.file_list.insert(END, file)

            self.file_list.bind('<<ListboxSelect>>', lambda event: self.open_file(self.file_list, folder_path))
            self.file_list.bind("<Motion>", lambda event: self.speak(self.file_list.get(self.file_list.nearest(event.y))))

            self.file_list.pack(fill=BOTH, expand=True)

        def open_file(self, file_list, folder_path):
            selected_file = file_list.get(file_list.curselection())
            file_path = os.path.join(folder_path, selected_file)

            if platform.system() == 'Darwin':       # macOS
                subprocess.call(('open', file_path))
            elif platform.system() == 'Windows':    # Windows
                os.startfile(file_path)
            else:                                   # linux variants
                subprocess.call(('xdg-open', file_path))

        def start_listening(self):
            with sr.Microphone() as source:
                audio = self.r.listen(source)
            try:
                command = self.r.recognize_google(audio)
                self.process_command(command)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))

        def process_command(self, command):
            # Process the command here
            # For example, if command == "Pictures", call self.display_file_type("Pictures")
            pass

        def quit_app(self):
            self.master.destroy()

        def toggle_speech(self):
            self.speaking = not self.speaking

        def speak(self, text):
            if self.speaking:
                self.engine.say(text)
                self.engine.runAndWait()
    
    def run_file_manager():
        root = Tk()
        app = FileManagerApp(root)
        root.mainloop()
    run_file_manager()
    
    


def run_calendar():
    class TalkingCalendarApp:
        def __init__(self):
            self.root = tk.Tk()
            self.root.title("Talking Calendar")
            self.root.geometry("800x600")

            self.cal = Calendar(self.root, selectmode='day',
                                year=datetime.now().year, month=datetime.now().month, day=datetime.now().day)
            self.cal.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)

            self.btn_select = tk.Button(self.root, text="Select Date", command=self.select_date)
            self.btn_select.pack(pady=10)

            self.announce_current_date()

            self.root.bind('<space>', self.on_space)

        def announce_current_date(self):
            today = datetime.today()
            date_str = today.strftime("%A, %B %d, %Y")
            self.speak(date_str)

        def select_date(self):
            chosen_date = self.cal.get_date()
            self.speak(f"You selected {chosen_date}")

            # NLP processing and event retrieval here

        def speak(self, text):
            engine = pyttsx3.init()
            engine.stop()
            engine.say(text)
            engine.runAndWait()

        def on_space(self, event):
            if event.keysym == 'space':
                self.select_date()

        def run_talking_calendar(self):
            self.root.mainloop()

    if __name__ == "__main__":
        app = TalkingCalendarApp()
        app.run_talking_calendar()


def run_stopwatch():
    class StopwatchApp:
        def __init__(self):
            self.counter = 66600
            self.running = False
            self.engine = pyttsx3.init()
            self.root = Tkinter.Tk()
            self.root.title("Stopwatch")

            # Set the window size and allow it to be resizable
            self.root.geometry("600x400")
            self.root.resizable(True, True)

            # Detect changes in window size
            self.root.bind("<Configure>", self.on_configure)

            self.label = Tkinter.Label(self.root, text="Stopwatch!", fg="black", font="Verdana 30 bold")
            self.label.pack()

            self.speak_instructions()

            self.f = Tkinter.Frame(self.root)
            self.start = Tkinter.Button(self.f, text='Start', width=6, command=self.start_stopwatch)
            self.stop = Tkinter.Button(self.f, text='Stop', width=6, state='disabled', command=self.stop_stopwatch)
            self.reset = Tkinter.Button(self.f, text='Reset', width=6, state='disabled', command=self.reset_stopwatch)
            self.quit_btn = Tkinter.Button(self.f, text='Quit', width=6, command=self.quit_app)

            self.f.pack(anchor='center', pady=5)
            self.start.pack(side="left")
            self.stop.pack(side="left")
            self.reset.pack(side="left")
            self.quit_btn.pack(side="left")

            self.root.bind("<s>", lambda event: self.start_stopwatch())
            self.root.bind("<space>", lambda event: self.stop_stopwatch())
            self.root.bind("<r>", lambda event: self.reset_stopwatch())
            self.root.bind("<q>", lambda event: self.quit_app())
            self.root.bind("<Return>", self.do_nothing)

            self.root.mainloop()

        def speak(self, text, rate=150):
            self.engine.setProperty('rate', rate)
            self.engine.say(text)
            self.engine.runAndWait()

        def speak_instructions(self):
            self.speak("Click on the 'S' key to start the Stopwatch. Click on the spacebar to stop the stopwatch. Click on the 'R' key to reset the stopwatch. Click on the 'Q' key to quit.", rate=170)

        def counter_label(self):
            def count():
                if self.running:
                    if self.counter == 66600:
                        display = "Starting..."
                        self.speak("Timer started")
                    else:
                        tt = datetime.fromtimestamp(self.counter)
                        string = tt.strftime("%H:%M:%S")
                        display = string
                    self.label['text'] = display
                    self.label.after(1000, count)
                    self.counter += 1
            count()

        def on_configure(self, event):
            self.root.geometry(self.root.geometry())

        def start_stopwatch(self):
            if not self.running:
                self.running = True
                self.counter_label()
                self.start['state'] = 'disabled'
                self.stop['state'] = 'normal'
                self.reset['state'] = 'normal'

        def stop_stopwatch(self):
            if self.running:
                tt = datetime.fromtimestamp(self.counter)
                string = tt.strftime("%H hours, %M minutes, and %S seconds")
                self.speak("Stopwatch stopped. The elapsed time is " + string + "click on s key again to continue the Stopwatch")
            self.start['state'] = 'normal'
            self.stop['state'] = 'disabled'
            self.reset['state'] = 'normal'
            self.running = False

        def reset_stopwatch(self):
            self.counter = 66600
            if not self.running:
                self.reset['state'] = 'disabled'
                self.label['text'] = 'Stopwatch!'
                self.speak("Timer reset")
            else:
                self.label['text'] = 'Starting...'

        def quit_app(self):
            self.root.destroy()

        def do_nothing(self, event):
            pass

    if __name__ == "__main__":
        app = StopwatchApp()

        

def run_speech_to_text_app():
    class SpeechToTextApp:
        def __init__(self, master):
            self.master = master
            master.title("Text Editor")

            # Set window size and position
            window_width = 800
            window_height = 600
            screen_width = master.winfo_screenwidth()
            screen_height = master.winfo_screenheight()
            x_coordinate = (screen_width - window_width) // 2
            y_coordinate = (screen_height - window_height) // 2
            master.geometry(f"{window_width}x{window_height}+{x_coordinate}+{y_coordinate}")

            self.text_output = scrolledtext.ScrolledText(master, wrap=tk.WORD)
            self.text_output.pack(fill=tk.BOTH, expand=True)

            self.start_button = tk.Button(master, text="Start Listening", command=self.listen_and_recognize)
            self.start_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.listen_button = tk.Button(master, text="Listen", command=self.read_text)
            self.listen_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.save_button = tk.Button(master, text="Save Text", command=self.save_text)
            self.save_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.exit_button = tk.Button(master, text="Exit", command=master.destroy)
            self.exit_button.pack(side=tk.RIGHT, padx=10, pady=5)

            # Bind keyboard events
            master.bind('<KeyPress-space>', lambda event: self.listen_and_recognize())
            master.bind('<Control-l>', lambda event: self.read_text())
            master.bind('<Control-s>', lambda event: self.save_text())
            master.bind('<KeyPress-q>', lambda event: master.destroy())

        def listen_and_recognize(self):
            recognizer = sr.Recognizer()

            with sr.Microphone() as source:
                print("Say something:")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)

            try:
                print("Recognizing...")
                text = recognizer.recognize_google(audio)
                self.text_output.insert(tk.END, f" {text}\n")
            except sr.UnknownValueError:
                print("Could not understand audio.")
                self.text_output.insert(tk.END, "Could not understand audio.\n")
            except sr.RequestError as e:
                print(f"Could not request results from Google Speech Recognition service; {e}")
                self.text_output.insert(tk.END, f"Error: {e}\n")

        def read_text(self):
            text = self.text_output.get("1.0", tk.END)
            if text.strip():  # Check if there is text to read
                engine = pyttsx3.init()
                engine.say(text)
                engine.runAndWait()

        def save_text(self):
            text = self.text_output.get("1.0", tk.END)
            if text.strip():  # Check if there is text to save
                with open("output.txt", "w") as file:
                    file.write(text)
                print("Text saved to 'output.txt'")

    if __name__ == "__main__":
        root = tk.Tk()
        app = SpeechToTextApp(root)
        root.mainloop()


def run_timer_app():
    class TimerApp:
        def __init__(self, root):
            self.root = root
            self.root.title("Modern Timer")

            self.hours_var = tk.StringVar(value="00")
            self.minutes_var = tk.StringVar(value="00")
            self.seconds_var = tk.StringVar(value="00")

            self.hour_label = tk.Label(root, text="Hour:")
            self.minute_label = tk.Label(root, text="Minute:")
            self.second_label = tk.Label(root, text="Second:")

            self.hour_entry = tk.Entry(root, textvariable=self.hours_var, width=3)
            self.minute_entry = tk.Entry(root, textvariable=self.minutes_var, width=3)
            self.second_entry = tk.Entry(root, textvariable=self.seconds_var, width=3)

            self.hour_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.hours_var))
            self.minute_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.minutes_var))
            self.second_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.seconds_var))

            self.start_button = tk.Button(root, text="Start Timer", command=self.start_timer)
            self.timer_label = tk.Label(root, text="", font=("Helvetica", 16))

            # Progress bar using a horizontal scale
            self.progress_bar = ttk.Progressbar(root, orient="horizontal", length=200, mode="determinate")
            self.progress_value = 0

            # Add button variables
            self.hour_button = tk.Button(root, text="Set Hours", command=self.set_hours)
            self.minute_button = tk.Button(root, text="Set Minutes", command=self.set_minutes)
            self.second_button = tk.Button(root, text="Set Seconds", command=self.set_seconds)

            self.engine = pyttsx3.init()

            self.setup_layout()
            self.bind_keys()

        def setup_layout(self):
            self.hour_label.pack(pady=5)
            self.hour_entry.pack(pady=5)

            self.minute_label.pack(pady=5)
            self.minute_entry.pack(pady=5)

            self.second_label.pack(pady=5)
            self.second_entry.pack(pady=5)

            self.start_button.pack(pady=10)
            self.timer_label.pack(pady=10)
            self.progress_bar.pack(pady=10)

            # Add buttons to the layout
            self.hour_button.pack(pady=5)
            self.minute_button.pack(pady=5)
            self.second_button.pack(pady=5)

        def bind_keys(self):
            self.root.bind("<Control-h>", lambda event: self.set_hours())
            self.root.bind("<Control-m>", lambda event: self.set_minutes())
            self.root.bind("<Control-s>", lambda event: self.set_seconds())
            self.root.bind("<space>", lambda event: self.start_timer())

        def get_time_values(self, limit):
            return [str(i).zfill(2) for i in range(limit)]

        def start_timer(self, event=None):
            try:
                hours = int(self.hours_var.get())
                minutes = int(self.minutes_var.get())
                seconds = int(self.seconds_var.get())

                total_seconds = hours * 3600 + minutes * 60 + seconds

                if total_seconds > 0:
                    self.hide_time_inputs()
                    self.show_timer_info(f"Timer set for {hours} hours, {minutes} minutes, and {seconds} seconds.")

                    self.progress_value = 0
                    self.progress_bar["value"] = self.progress_value
                    self.progress_bar["maximum"] = total_seconds

                    self.engine.say("Timer started")
                    self.engine.runAndWait()

                    while self.progress_value < total_seconds:
                        self.progress_value += 1
                        mins, secs = divmod(total_seconds - self.progress_value, 60)
                        time_format = "{:02d}:{:02d}".format(mins, secs)
                        self.timer_label.config(text=time_format)
                        self.progress_bar["value"] = self.progress_value
                        self.root.update()
                        time.sleep(1)

                    self.engine.say("Time over")
                    self.engine.runAndWait()

                    self.show_timer_info("Timer completed!")
                    self.show_time_inputs()
                else:
                    messagebox.showwarning("Invalid Input", "Please set a positive time.")
            except ValueError:
                messagebox.showwarning("Invalid Input", "Please enter valid numbers for hours, minutes, and seconds.")

        def show_timer_info(self, message):
            messagebox.showinfo("Timer", message)

        def set_hours(self, event=None):
            user_input = self.get_user_input("Set Hours", "Enter the number of hours:")
            if user_input is not None:
                self.hours_var.set(str(user_input).zfill(2))

        def set_minutes(self, event=None):
            user_input = self.get_user_input("Set Minutes", "Enter the number of minutes:")
            if user_input is not None:
                self.minutes_var.set(str(user_input).zfill(2))

        def set_seconds(self, event=None):
            user_input = self.get_user_input("Set Seconds", "Enter the number of seconds:")
            if user_input is not None:
                self.seconds_var.set(str(user_input).zfill(2))

        def get_user_input(self, title, prompt):
            return simpledialog.askinteger(title, prompt, parent=self.root)

        def hide_time_inputs(self):
            self.hour_label.pack_forget()
            self.hour_entry.pack_forget()
            self.minute_label.pack_forget()
            self.minute_entry.pack_forget()
            self.second_label.pack_forget()
            self.second_entry.pack_forget()

        def show_time_inputs(self):
            self.hour_label.pack(pady=5)
            self.hour_entry.pack(pady=5)
            self.minute_label.pack(pady=5)
            self.minute_entry.pack(pady=5)
            self.second_label.pack(pady=5)
            self.second_entry.pack(pady=5)

        def clear_default_value(self, var):
            if var.get() == "00":
                var.set("")

    if __name__ == "__main__":
        root = tk.Tk()
        app = TimerApp(root)
        root.mainloop()





def run_camera():
    class CameraApp:
        def __init__(self):
            # Initialize the pyttsx3 engine
            self.engine = pyttsx3.init()

            # Get a video capture object for the camera (usually default webcam)
            self.vid = cv2.VideoCapture(0)

            # Specify the directory to save captured photos
            self.photos_directory = "C:\\Users\\Asus\\OneDrive\\Pictures"

            # Specify the directory to save the recorded video
            self.videos_directory = "C:\\Users\\Asus\\OneDrive\\Pictures"

            # Create the directories if they don't exist
            os.makedirs(self.photos_directory, exist_ok=True)
            os.makedirs(self.videos_directory, exist_ok=True)

            # Variable to track if video recording is in progress
            self.recording = False

            # Variable to track if a face has been detected
            self.face_detected = False

            # Variable to track if the face detection announcement has been made
            self.face_detection_announced = False

            # Define audio parameters
            self.FORMAT = pyaudio.paInt16
            self.CHANNELS = 1
            self.RATE = 44100
            self.CHUNK = 1024

            # Load the pre-trained face detection classifier
            self.face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

            # Create a resizable window
            cv2.namedWindow('Camera Feed', cv2.WINDOW_NORMAL)
            cv2.setMouseCallback('Camera Feed', self.resize_window)

            while True:
                # Read frames from the camera
                ret, frame = self.vid.read()

                # Convert the frame to grayscale for face detection
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

                # Detect faces in the frame
                faces = self.face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

                # If a face is detected and it hasn't been detected before, announce it
                if len(faces) > 0 and not self.face_detection_announced and not self.recording:  # Add 'not recording' condition
                    self.speak_message("Face detected")
                    self.face_detection_announced = True
                elif len(faces) == 0:
                    self.face_detection_announced = False

                # Show the frames in a window
                cv2.imshow('Camera Feed', frame)

                # Add a button to capture a photo when 'c' key is pressed
                # and start/stop video recording when 'v' key is pressed
                key = cv2.waitKey(1) & 0xFF
                if key == ord('c'):
                    self.save_photo(frame)
                elif key == ord('v'):
                    if not self.recording:
                        # Start video recording
                        self.speak_message("Video recording started. Press 'v' again to stop.")
                        self.start_video_recording(frame)
                    else:
                        # Stop video recording
                        self.stop_video_recording()
                elif key == ord('q'):
                    break

                # Write the frame to the output video file when recording is in progress
                if self.recording:
                    self.output_video.write(frame)

            # Release the video capture object and close all windows
            self.vid.release()
            cv2.destroyAllWindows()

        def save_photo(self, frame):
            # Save the captured photo with a unique filename
            filename = os.path.join(self.photos_directory, f"photo_{len(os.listdir(self.photos_directory)) + 1}.jpg")
            cv2.imwrite(filename, frame)
            print(f"Photo saved as {filename}")
            # Speak out that the photo has been saved
            self.speak_message("Photo saved")

        def speak_message(self, message):
            # Speak out the provided message
            self.engine.say(message)
            self.engine.runAndWait()

        def record_audio(self, output_filename):
            audio = pyaudio.PyAudio()
            stream = audio.open(format=self.FORMAT,
                                channels=self.CHANNELS,
                                rate=self.RATE,
                                input=True,
                                frames_per_buffer=self.CHUNK)
            frames = []
            print("Recording audio...")
            while self.recording:
                data = stream.read(self.CHUNK)
                frames.append(data)
            print("Finished recording audio.")

            # Stop and close the audio stream
            stream.stop_stream()
            stream.close()
            audio.terminate()

            # Save the recorded audio to a WAV file
            wf = wave.open(output_filename, 'wb')
            wf.setnchannels(self.CHANNELS)
            wf.setsampwidth(audio.get_sample_size(self.FORMAT))
            wf.setframerate(self.RATE)
            wf.writeframes(b''.join(frames))
            wf.close()

        def start_audio_recording(self, output_filename):
            self.recording = True
            self.record_audio(output_filename)

        def stop_audio_recording(self):
            self.recording = False

        def merge_audio_video(self, video_filename, audio_filename, output_filename):
            try:
                video_clip = VideoFileClip(video_filename)
                audio_clip = AudioFileClip(audio_filename)
                video_clip = video_clip.set_audio(audio_clip)
                video_clip.write_videofile(output_filename, codec='libx264', audio_codec='aac')
                video_clip.close()
                audio_clip.close()
            except Exception as e:
                print(f"Error merging audio and video: {e}")

        def start_video_recording(self, frame):
            # Start video recording
            self.recording = True
            # Define the codec and create a VideoWriter object to save the video
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            self.output_video_filename = os.path.join(self.videos_directory, "my_gallery_video.avi")
            self.output_video = cv2.VideoWriter(self.output_video_filename, fourcc, 20.0, (640, 480))
            # Start audio recording in a separate thread
            self.audio_filename = os.path.join(self.videos_directory, "audio.wav")
            audio_thread = threading.Thread(target=self.start_audio_recording, args=(self.audio_filename,))
            audio_thread.start()
            # Update face_detection_announced to False when recording starts
            self.face_detection_announced = False

        def stop_video_recording(self):
            # Stop video recording
            self.speak_message("Video recording stopped.")
            self.output_video.release()
            self.stop_audio_recording()
            self.audio_thread.join()
            # Merge audio and video files
            merged_video_filename = os.path.join(self.videos_directory, "merged_video.mp4")
            self.merge_audio_video(self.output_video_filename, self.audio_filename, merged_video_filename)

        def resize_window(self, event, x, y, flags, param):
            if event == cv2.EVENT_MAXIMIZED:
                cv2.setWindowProperty('Camera Feed', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

    if __name__ == "__main__":
        app = CameraApp()


        
    
def main_frame_window():
    root = tk.Tk()
    root.title("Main Frame Window")

#    timer_button = tk.Button(root, text="Open Timer App", command=open_timer_app)
#    timer_button.pack(pady=10)

#    speech_to_text_button = tk.Button(root, text="Open Speech to Text App", command=open_speech_to_text_app)
#    speech_to_text_button.pack(pady=10)

    calculator_button = tk.Button(root, text="Open Calculator App", command=run_calculator)
    calculator_button.pack(pady=10)

    web_browser_button = tk.Button(root, text="Open Web Browser App", command=run_web_browser)
    web_browser_button.pack(pady=10)

    file_manager_button = tk.Button(root, text="File Manager App", command=run_file_manager)
    file_manager_button.pack(pady=10)
    
    calendar_button = tk.Button(root, text="Calendar", command=run_calendar)
    calendar_button.pack(pady=10)
    
    stopwatch_button = tk.Button(root, text="Stopwatch", command=run_stopwatch)
    stopwatch_button.pack(pady=10)
    
    speech_to_text_button = tk.Button(root, text="Text Editor App", command=run_speech_to_text_app)
    speech_to_text_button.pack(pady=10)

    timer_button = tk.Button(root, text="Timer App", command=run_timer_app)
    timer_button.pack(pady=10)
    
    camera_button = tk.Button(root, text="Camera", command=run_camera)
    camera_button.pack(pady=10)
    
    root.mainloop()



if __name__ == "__main__":
    main_frame_window()
```

```python
# final code for All applications.....................................................................
import tkinter as tk
from tkinter import messagebox, scrolledtext
import time
import pyttsx3
import speech_recognition as sr
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QToolBar, QAction, QLineEdit
from PyQt5.QtCore import QUrl
from PyQt5.QtWebEngineWidgets import QWebEngineView
from PyQt5.QtGui import QKeySequence
from PyQt5.QtCore import Qt
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog
import os
from PIL import Image, ImageTk
import tkinter as tk
from datetime import datetime
from tkcalendar import Calendar
import pyttsx3
import tkinter as Tkinter
from datetime import datetime
import pyttsx3
import tkinter as tk
from tkinter import scrolledtext
import speech_recognition as sr
import pyttsx3
import tkinter as tk
from tkinter import messagebox
import time
import pyttsx3
import tkinter as tk
import speech_recognition as sr
import pyttsx3
import tkinter as Tkinter
from datetime import datetime
import pyttsx3
import tkinter as tk
from tkinter import scrolledtext
import speech_recognition as sr
import pyttsx3
import cv2
import os
import pyttsx3
import pyaudio
import wave
import threading
from moviepy.editor import VideoFileClip, AudioFileClip
from tkinter import *
from PIL import Image, ImageTk
import os
import speech_recognition as sr
import pyttsx3
import keyboard
import subprocess
import platform
import tkinter as tk
from tkinter import ttk, simpledialog
from tkinter import messagebox
import time
import pyttsx3


def run_calculator():
    class Calculator:
        def __init__(self, master):
            self.master = master
            self.master.title('Calculator')
            self.equation = ""
            self.speech_engine = pyttsx3.init()
            self.recognizer = sr.Recognizer()
            self.stop_listening = None
            self.is_listening = False

            # Entry field to show calculations
            self.entry_field = tk.Entry(master, width=40, borderwidth=5, font=('arial', 20, 'bold'))
            self.entry_field.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky='we')

            # Buttons
            buttons = [
                '7', '8', '9', '/',
                '4', '5', '6', '*',
                '1', '2', '3', '-',
                '0', '.', '=', '+'
            ]

            row_val = 1
            col_val = 0

            for button in buttons:
                btn = tk.Button(master, text=button, width=10, font=('arial', 20, 'bold'), command=lambda button=button: self.click_button(button))
                btn.grid(row=row_val, column=col_val, sticky='we')
                btn.bind('<Enter>', lambda event, button=button: self.speak_button(event, button))
                col_val += 1
                if col_val > 3:
                    col_val = 0
                    row_val += 1

            # Clear and backspace buttons
            clear_button = tk.Button(master, text='Clear', width=10, font=('arial', 20, 'bold'), command=self.clear_all)
            clear_button.grid(row=row_val, column=0, columnspan=2, sticky='we')
            clear_button.bind('<Enter>', self.speak_clear)

            backspace_button = tk.Button(master, text='Backspace', width=10, font=('arial', 20, 'bold'), command=self.backspace)
            backspace_button.grid(row=row_val, column=2, columnspan=2, sticky='we')
            backspace_button.bind('<Enter>', self.speak_backspace)

            # Quit button
            quit_button = tk.Button(master, text='Quit', width=10, font=('arial', 20, 'bold'), command=self.quit_app)
            quit_button.grid(row=row_val+1, column=0, columnspan=4, sticky='we')
            quit_button.bind('<Enter>', self.speak_quit)

            # Bind keys to start and stop listening
            master.bind('<Control-Shift-KeyPress-V>', self.start_listening)
            master.bind('<Control-Shift-KeyRelease-V>', self.stop_listening_and_calculate)
            master.bind('<BackSpace>', self.backspace_input)
            master.bind('<Delete>', self.clear_all_input)
            master.bind('<Key>', self.keyboard_input)
            master.bind('<Return>', self.calculate_result)

        def click_button(self, button):
            if button == '=':
                # Perform calculation
                try:
                    self.equation = str(eval(self.equation))
                    self.speech_engine.say("The result is " + self.equation)
                    self.speech_engine.runAndWait()
                except:
                    self.equation = "Error"
                    self.speech_engine.say("Error")
                    self.speech_engine.runAndWait()
            else:
                # Add button press to equation
                self.equation += str(button)
                self.speech_engine.say(str(button))
                self.speech_engine.runAndWait()
            self.update_entry_field()

        def clear_all(self):
            self.equation = ""
            self.update_entry_field()
            self.speech_engine.say("Clear")
            self.speech_engine.runAndWait()

        def backspace(self):
            self.equation = self.equation[:-1]
            self.update_entry_field()
            self.speech_engine.say("Backspace")
            self.speech_engine.runAndWait()

        def speak_button(self, event, button):
            self.speech_engine.say(button)
            self.speech_engine.runAndWait()

        def speak_clear(self, event):
            self.speech_engine.say("Clear")
            self.speech_engine.runAndWait()

        def speak_backspace(self, event):
            self.speech_engine.say("Backspace")
            self.speech_engine.runAndWait()

        def speak_quit(self, event):
            self.speech_engine.say("Quit")
            self.speech_engine.runAndWait()

        def start_listening(self, event):
            if not self.is_listening:
                self.is_listening = True
                with sr.Microphone() as source:
                    audio = self.recognizer.listen(source)
                    try:
                        command = self.recognizer.recognize_google(audio)
                        self.equation += command
                        self.update_entry_field()
                    except sr.UnknownValueError:
                        pass
                    except sr.RequestError:
                        pass

        def stop_listening_and_calculate(self, event):
            if self.is_listening:
                self.is_listening = False
                self.click_button('=')

        def clear_all_input(self, event):
            self.clear_all()

        def backspace_input(self, event):
            self.backspace()

        def update_entry_field(self):
            self.entry_field.delete(0, tk.END)
            self.entry_field.insert(0, self.equation)
            self.speech_engine.say(self.equation)
            self.speech_engine.runAndWait()

        def keyboard_input(self, event):
            key = event.keysym
            if key.isdigit() or key in ['+', '-', '*', '/', '.', '=', 'BackSpace', 'Delete']:
                if key == 'BackSpace':
                    self.backspace()
                elif key == 'Delete':
                    self.clear_all()
                elif key == '=':
                    self.click_button('=')
                else:
                    self.click_button(key)
            elif key == 'plus':
                self.click_button('+')
            elif key == 'minus':
                self.click_button('-')
            elif key == 'asterisk':
                self.click_button('*')
            elif key == 'slash':
                self.click_button('/')
            else:
                pass

        def calculate_result(self, event):
            self.click_button('=')

        def quit_app(self):
            self.speech_engine.say("Quit")
            self.speech_engine.runAndWait()
            self.master.quit()

    root = tk.Tk()
    root.state('zoomed')  # This line makes the window maximized
    root.grid_columnconfigure(0, weight=1)  # These lines make the widgets expand when the window size changes
    root.grid_columnconfigure(1, weight=1)
    root.grid_columnconfigure(2, weight=1)
    root.grid_columnconfigure(3, weight=1)
    root.grid_rowconfigure(0, weight=1)
    root.grid_rowconfigure(1, weight=1)
    root.grid_rowconfigure(2, weight=1)
    root.grid_rowconfigure(3, weight=1)
    root.grid_rowconfigure(4, weight=1)
    root.grid_rowconfigure(5, weight=1)
    my_calculator = Calculator(root)
    root.mainloop()

# Call the function to run the calculator
run_calculator()



def run_web_browser():
    class WebBrowser(QMainWindow):
        def __init__(self):
            super().__init__()

            # Initialize the web browser view
            self.browser = QWebEngineView()
            self.browser.setUrl(QUrl("http://www.google.com"))

            # Set the web browser as the central widget
            self.setCentralWidget(self.browser)

            # Create navigation bar
            navbar = QToolBar()
            self.addToolBar(navbar)

            # Back Button
            back_btn = QAction('Back', self)
            back_btn.setStatusTip('Back to previous page')
            back_btn.triggered.connect(self.browser.back)
            navbar.addAction(back_btn)

            # Forward Button
            forward_btn = QAction('Forward', self)
            forward_btn.setStatusTip('Forward to next page')
            forward_btn.triggered.connect(self.browser.forward)
            navbar.addAction(forward_btn)

            # Reload Button
            reload_btn = QAction('Reload', self)
            reload_btn.setStatusTip('Reload page')
            reload_btn.triggered.connect(self.browser.reload)
            navbar.addAction(reload_btn)

            # URL Bar
            self.url_bar = QLineEdit()
            self.url_bar.returnPressed.connect(self.navigate_to_url)
            navbar.addWidget(self.url_bar)

            # Home Button
            home_btn = QAction('Home', self)
            home_btn.setStatusTip('Go home')
            home_btn.triggered.connect(self.navigate_home)
            navbar.addAction(home_btn)

            # Stop Button
            stop_btn = QAction('Stop', self)
            stop_btn.setStatusTip('Stop loading current page')
            stop_btn.triggered.connect(self.browser.stop)
            navbar.addAction(stop_btn)

            # URL Changed Event
            self.browser.urlChanged.connect(self.update_urlbar)

            # Go Back, Go Forward, Refresh, URL Bar Context Menu
            navbar.addSeparator()

            # Go Back Button
            back_btn = QAction('Go Back', self)
            back_btn.triggered.connect(self.browser.back)
            navbar.addAction(back_btn)

            # Go Forward Button
            forward_btn = QAction('Go Forward', self)
            forward_btn.triggered.connect(self.browser.forward)
            navbar.addAction(forward_btn)

            # Refresh Button
            refresh_btn = QAction('Refresh', self)
            refresh_btn.triggered.connect(self.browser.reload)
            navbar.addAction(refresh_btn)

            # Clear URL Bar Button
            clear_url_btn = QAction('Clear URL', self)
            clear_url_btn.triggered.connect(self.clear_urlbar)
            navbar.addAction(clear_url_btn)

            # Voice Command Activation Button
            voice_activation_btn = QAction('Voice Command (Space)', self)
            voice_activation_btn.setShortcut(QKeySequence(Qt.Key_Space))
            voice_activation_btn.triggered.connect(self.activate_voice_command)
            self.addAction(voice_activation_btn)

            # Initialize speech recognizer
            self.recognizer = sr.Recognizer()
            self.microphone = sr.Microphone()

        def activate_voice_command(self):
            with self.microphone as source:
                self.recognizer.adjust_for_ambient_noise(source)
                print("Listening...")
                audio = self.recognizer.listen(source)

            try:
                command = self.recognizer.recognize_google(audio).lower()
                print("Command:", command)
                self.process_voice_command(command)
            except sr.UnknownValueError:
                print("Could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))

        def process_voice_command(self, command):
            # Define commands and their corresponding actions here
            if 'go back' in command:
                self.browser.back()
            elif 'go forward' in command:
                self.browser.forward()
            elif 'refresh' in command:
                self.browser.reload()
            elif 'stop' in command:
                self.browser.stop()
            elif 'navigate' in command:
                # Extract URL from command and navigate
                url = command.split('navigate to ')[1]
                self.navigate_to_url(url)
            elif 'home' in command:
                self.navigate_home()

        def navigate_home(self):
            self.browser.setUrl(QUrl("http://www.google.com"))

        def navigate_to_url(self, url=None):
            if not url:
                url = self.url_bar.text()
            q = QUrl(url)
            if q.scheme() == "":
                q.setScheme("http")
            self.browser.setUrl(q)

        def update_urlbar(self, q):
            self.url_bar.setText(q.toString())
            self.url_bar.setCursorPosition(0)

        def clear_urlbar(self):
            self.url_bar.clear()

    app = QApplication(sys.argv)
    QApplication.setApplicationName("Python Web Browser")

    main = WebBrowser()
    main.show()

    app.exec_()

    
#final code for filemanager


def run_file_manager():
    class FileManagerApp:
        def __init__(self, master):
            self.master = master
            master.title("File Manager")

            self.header_frame = Frame(master)
            self.header_frame.pack(side=TOP, fill=X)

            self.back_button = Button(self.header_frame, text="Back", command=self.display_file_manager, width=20, height=1)
            self.back_button.pack(side=LEFT)

            self.button_frame = Frame(master)
            self.button_frame.pack(side=LEFT, fill=Y)

            self.listbox_frame = Frame(master)
            self.listbox_frame.pack(side=RIGHT, fill=BOTH, expand=True)

            self.display_file_manager()

            # Bind arrow keys to operate buttons
            master.bind("<Up>", lambda event: self.navigate_buttons(-1))
            master.bind("<Down>", lambda event: self.navigate_buttons(1))
            master.bind("<Left>", lambda event: self.navigate_files(-1))
            master.bind("<Right>", lambda event: self.navigate_files(1))
            master.bind("<Return>", lambda event: self.select_button())

            # Initialize the recognizer and the speech engine
            self.r = sr.Recognizer()
            self.engine = pyttsx3.init()

            # Add these lines to the __init__ method of FileManagerApp
            self.speaking = False
            keyboard.add_hotkey('space', self.start_listening)
            keyboard.add_hotkey('q', self.quit_app)
            keyboard.add_hotkey('ctrl + s', self.toggle_speech)

        def display_file_manager(self):
            for widget in self.button_frame.winfo_children():
                widget.destroy()

            for widget in self.listbox_frame.winfo_children():
                widget.destroy()

            picture_button = Button(self.button_frame, text="Pictures", command=lambda: self.display_file_type("Pictures"), width=20, height=1)
            video_button = Button(self.button_frame, text="Videos", command=lambda: self.display_file_type("Videos"), width=20, height=1)
            document_button = Button(self.button_frame, text="Documents", command=lambda: self.display_file_type("Documents"), width=20, height=1)
            music_button = Button(self.button_frame, text="Music", command=lambda: self.display_file_type("Music"), width=20, height=1)
            download_button = Button(self.button_frame, text="Download", command=lambda: self.display_file_type("Download"), width=20, height=1)

            self.buttons = [picture_button, video_button, document_button, music_button, download_button]
            self.selected_button_index = 0
            self.buttons[self.selected_button_index].config(highlightbackground="blue")

            for button in self.buttons:
                button.bind("<Enter>", lambda event, text=button.cget("text"): self.speak(text))
                button.pack(fill=Y, expand=True)

        def navigate_buttons(self, direction):
            self.buttons[self.selected_button_index].config(highlightbackground="SystemButtonFace")
            self.selected_button_index = (self.selected_button_index + direction) % len(self.buttons)
            self.buttons[self.selected_button_index].config(highlightbackground="blue")
            self.speak(self.buttons[self.selected_button_index].cget("text"))

        def navigate_files(self, direction):
            if hasattr(self, 'file_list'):
                current_selection = self.file_list.curselection()
                if current_selection:
                    self.file_list.selection_clear(current_selection)
                    new_index = (current_selection[0] + direction) % self.file_list.size()
                    self.file_list.selection_set(new_index)
                    self.file_list.see(new_index)
                    self.speak(self.file_list.get(new_index))

        def select_button(self):
            self.buttons[self.selected_button_index].invoke()

        def display_file_type(self, folder_name):
            for widget in self.listbox_frame.winfo_children():
                widget.destroy()

            self.file_list = Listbox(self.listbox_frame)

            home_dir = os.path.expanduser("~")
            folder_path = os.path.join(home_dir, folder_name)

            for file in os.listdir(folder_path):
                self.file_list.insert(END, file)

            self.file_list.bind('<<ListboxSelect>>', lambda event: self.open_file(self.file_list, folder_path))
            self.file_list.bind("<Motion>", lambda event: self.speak(self.file_list.get(self.file_list.nearest(event.y))))

            self.file_list.pack(fill=BOTH, expand=True)

        def open_file(self, file_list, folder_path):
            selected_file = file_list.get(file_list.curselection())
            file_path = os.path.join(folder_path, selected_file)

            if platform.system() == 'Darwin':       # macOS
                subprocess.call(('open', file_path))
            elif platform.system() == 'Windows':    # Windows
                os.startfile(file_path)
            else:                                   # linux variants
                subprocess.call(('xdg-open', file_path))

        def start_listening(self):
            with sr.Microphone() as source:
                audio = self.r.listen(source)
            try:
                command = self.r.recognize_google(audio)
                self.process_command(command)
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))

        def process_command(self, command):
            # Process the command here
            # For example, if command == "Pictures", call self.display_file_type("Pictures")
            pass

        def quit_app(self):
            self.master.destroy()

        def toggle_speech(self):
            self.speaking = not self.speaking

        def speak(self, text):
            if self.speaking:
                self.engine.say(text)
                self.engine.runAndWait()
    
    def run_file_manager():
        root = Tk()
        app = FileManagerApp(root)
        root.mainloop()
    run_file_manager()
    
    


def run_calendar():
    class TalkingCalendarApp:
        def __init__(self):
            self.root = tk.Tk()
            self.root.title("Talking Calendar")
            self.root.geometry("800x600")

            self.cal = Calendar(self.root, selectmode='day',
                                year=datetime.now().year, month=datetime.now().month, day=datetime.now().day)
            self.cal.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)

            self.btn_select = tk.Button(self.root, text="Select Date", command=self.select_date)
            self.btn_select.pack(pady=10)

            self.announce_current_date()

            self.root.bind('<space>', self.on_space)

        def announce_current_date(self):
            today = datetime.today()
            date_str = today.strftime("%A, %B %d, %Y")
            self.speak(date_str)

        def select_date(self):
            chosen_date = self.cal.get_date()
            self.speak(f"You selected {chosen_date}")

            # NLP processing and event retrieval here

        def speak(self, text):
            engine = pyttsx3.init()
            engine.stop()
            engine.say(text)
            engine.runAndWait()

        def on_space(self, event):
            if event.keysym == 'space':
                self.select_date()

        def run_talking_calendar(self):
            self.root.mainloop()

    if __name__ == "__main__":
        app = TalkingCalendarApp()
        app.run_talking_calendar()


def run_stopwatch():
    class StopwatchApp:
        def __init__(self):
            self.counter = 66600
            self.running = False
            self.engine = pyttsx3.init()
            self.root = Tkinter.Tk()
            self.root.title("Stopwatch")

            # Set the window size and allow it to be resizable
            self.root.geometry("600x400")
            self.root.resizable(True, True)

            # Detect changes in window size
            self.root.bind("<Configure>", self.on_configure)

            self.label = Tkinter.Label(self.root, text="Stopwatch!", fg="black", font="Verdana 30 bold")
            self.label.pack()

            self.speak_instructions()

            self.f = Tkinter.Frame(self.root)
            self.start = Tkinter.Button(self.f, text='Start', width=6, command=self.start_stopwatch)
            self.stop = Tkinter.Button(self.f, text='Stop', width=6, state='disabled', command=self.stop_stopwatch)
            self.reset = Tkinter.Button(self.f, text='Reset', width=6, state='disabled', command=self.reset_stopwatch)
            self.quit_btn = Tkinter.Button(self.f, text='Quit', width=6, command=self.quit_app)

            self.f.pack(anchor='center', pady=5)
            self.start.pack(side="left")
            self.stop.pack(side="left")
            self.reset.pack(side="left")
            self.quit_btn.pack(side="left")

            self.root.bind("<s>", lambda event: self.start_stopwatch())
            self.root.bind("<space>", lambda event: self.stop_stopwatch())
            self.root.bind("<r>", lambda event: self.reset_stopwatch())
            self.root.bind("<q>", lambda event: self.quit_app())
            self.root.bind("<Return>", self.do_nothing)

            self.root.mainloop()

        def speak(self, text, rate=150):
            self.engine.setProperty('rate', rate)
            self.engine.say(text)
            self.engine.runAndWait()

        def speak_instructions(self):
            self.speak("Click on the 'S' key to start the Stopwatch. Click on the spacebar to stop the stopwatch. Click on the 'R' key to reset the stopwatch. Click on the 'Q' key to quit.", rate=170)

        def counter_label(self):
            def count():
                if self.running:
                    if self.counter == 66600:
                        display = "Starting..."
                        self.speak("Timer started")
                    else:
                        tt = datetime.fromtimestamp(self.counter)
                        string = tt.strftime("%H:%M:%S")
                        display = string
                    self.label['text'] = display
                    self.label.after(1000, count)
                    self.counter += 1
            count()

        def on_configure(self, event):
            self.root.geometry(self.root.geometry())

        def start_stopwatch(self):
            if not self.running:
                self.running = True
                self.counter_label()
                self.start['state'] = 'disabled'
                self.stop['state'] = 'normal'
                self.reset['state'] = 'normal'

        def stop_stopwatch(self):
            if self.running:
                tt = datetime.fromtimestamp(self.counter)
                string = tt.strftime("%H hours, %M minutes, and %S seconds")
                self.speak("Stopwatch stopped. The elapsed time is " + string + "click on s key again to continue the Stopwatch")
            self.start['state'] = 'normal'
            self.stop['state'] = 'disabled'
            self.reset['state'] = 'normal'
            self.running = False

        def reset_stopwatch(self):
            self.counter = 66600
            if not self.running:
                self.reset['state'] = 'disabled'
                self.label['text'] = 'Stopwatch!'
                self.speak("Timer reset")
            else:
                self.label['text'] = 'Starting...'

        def quit_app(self):
            self.root.destroy()

        def do_nothing(self, event):
            pass

    if __name__ == "__main__":
        app = StopwatchApp()

        

def run_speech_to_text_app():
    class SpeechToTextApp:
        def __init__(self, master):
            self.master = master
            master.title("Text Editor")

            # Set window size and position
            window_width = 800
            window_height = 600
            screen_width = master.winfo_screenwidth()
            screen_height = master.winfo_screenheight()
            x_coordinate = (screen_width - window_width) // 2
            y_coordinate = (screen_height - window_height) // 2
            master.geometry(f"{window_width}x{window_height}+{x_coordinate}+{y_coordinate}")

            self.text_output = scrolledtext.ScrolledText(master, wrap=tk.WORD)
            self.text_output.pack(fill=tk.BOTH, expand=True)

            self.start_button = tk.Button(master, text="Start Listening", command=self.listen_and_recognize)
            self.start_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.listen_button = tk.Button(master, text="Listen", command=self.read_text)
            self.listen_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.save_button = tk.Button(master, text="Save Text", command=self.save_text)
            self.save_button.pack(side=tk.LEFT, padx=10, pady=5)

            self.exit_button = tk.Button(master, text="Exit", command=master.destroy)
            self.exit_button.pack(side=tk.RIGHT, padx=10, pady=5)

            # Bind keyboard events
            master.bind('<KeyPress-space>', lambda event: self.listen_and_recognize())
            master.bind('<Control-l>', lambda event: self.read_text())
            master.bind('<Control-s>', lambda event: self.save_text())
            master.bind('<KeyPress-q>', lambda event: master.destroy())

        def listen_and_recognize(self):
            recognizer = sr.Recognizer()

            with sr.Microphone() as source:
                print("Say something:")
                recognizer.adjust_for_ambient_noise(source)  # Adjust for ambient noise
                audio = recognizer.listen(source)

            try:
                print("Recognizing...")
                text = recognizer.recognize_google(audio)
                self.text_output.insert(tk.END, f" {text}\n")
            except sr.UnknownValueError:
                print("Could not understand audio.")
                self.text_output.insert(tk.END, "Could not understand audio.\n")
            except sr.RequestError as e:
                print(f"Could not request results from Google Speech Recognition service; {e}")
                self.text_output.insert(tk.END, f"Error: {e}\n")

        def read_text(self):
            text = self.text_output.get("1.0", tk.END)
            if text.strip():  # Check if there is text to read
                engine = pyttsx3.init()
                engine.say(text)
                engine.runAndWait()

        def save_text(self):
            text = self.text_output.get("1.0", tk.END)
            if text.strip():  # Check if there is text to save
                with open("output.txt", "w") as file:
                    file.write(text)
                print("Text saved to 'output.txt'")

    if __name__ == "__main__":
        root = tk.Tk()
        app = SpeechToTextApp(root)
        root.mainloop()


def run_timer_app():
    class TimerApp:
        def __init__(self, root):
            self.root = root
            self.root.title("Modern Timer")

            self.hours_var = tk.StringVar(value="00")
            self.minutes_var = tk.StringVar(value="00")
            self.seconds_var = tk.StringVar(value="00")

            self.hour_label = tk.Label(root, text="Hour:")
            self.minute_label = tk.Label(root, text="Minute:")
            self.second_label = tk.Label(root, text="Second:")

            self.hour_entry = tk.Entry(root, textvariable=self.hours_var, width=3)
            self.minute_entry = tk.Entry(root, textvariable=self.minutes_var, width=3)
            self.second_entry = tk.Entry(root, textvariable=self.seconds_var, width=3)

            self.hour_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.hours_var))
            self.minute_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.minutes_var))
            self.second_entry.bind("<FocusIn>", lambda event: self.clear_default_value(self.seconds_var))

            self.start_button = tk.Button(root, text="Start Timer", command=self.start_timer)
            self.timer_label = tk.Label(root, text="", font=("Helvetica", 16))

            # Progress bar using a horizontal scale
            self.progress_bar = ttk.Progressbar(root, orient="horizontal", length=200, mode="determinate")
            self.progress_value = 0

            # Add button variables
            self.hour_button = tk.Button(root, text="Set Hours", command=self.set_hours)
            self.minute_button = tk.Button(root, text="Set Minutes", command=self.set_minutes)
            self.second_button = tk.Button(root, text="Set Seconds", command=self.set_seconds)

            self.engine = pyttsx3.init()

            self.setup_layout()
            self.bind_keys()

        def setup_layout(self):
            self.hour_label.pack(pady=5)
            self.hour_entry.pack(pady=5)

            self.minute_label.pack(pady=5)
            self.minute_entry.pack(pady=5)

            self.second_label.pack(pady=5)
            self.second_entry.pack(pady=5)

            self.start_button.pack(pady=10)
            self.timer_label.pack(pady=10)
            self.progress_bar.pack(pady=10)

            # Add buttons to the layout
            self.hour_button.pack(pady=5)
            self.minute_button.pack(pady=5)
            self.second_button.pack(pady=5)

        def bind_keys(self):
            self.root.bind("<Control-h>", lambda event: self.set_hours())
            self.root.bind("<Control-m>", lambda event: self.set_minutes())
            self.root.bind("<Control-s>", lambda event: self.set_seconds())
            self.root.bind("<space>", lambda event: self.start_timer())

        def get_time_values(self, limit):
            return [str(i).zfill(2) for i in range(limit)]

        def start_timer(self, event=None):
            try:
                hours = int(self.hours_var.get())
                minutes = int(self.minutes_var.get())
                seconds = int(self.seconds_var.get())

                total_seconds = hours * 3600 + minutes * 60 + seconds

                if total_seconds > 0:
                    self.hide_time_inputs()
                    self.show_timer_info(f"Timer set for {hours} hours, {minutes} minutes, and {seconds} seconds.")

                    self.progress_value = 0
                    self.progress_bar["value"] = self.progress_value
                    self.progress_bar["maximum"] = total_seconds

                    self.engine.say("Timer started")
                    self.engine.runAndWait()

                    while self.progress_value < total_seconds:
                        self.progress_value += 1
                        mins, secs = divmod(total_seconds - self.progress_value, 60)
                        time_format = "{:02d}:{:02d}".format(mins, secs)
                        self.timer_label.config(text=time_format)
                        self.progress_bar["value"] = self.progress_value
                        self.root.update()
                        time.sleep(1)

                    self.engine.say("Time over")
                    self.engine.runAndWait()

                    self.show_timer_info("Timer completed!")
                    self.show_time_inputs()
                else:
                    messagebox.showwarning("Invalid Input", "Please set a positive time.")
            except ValueError:
                messagebox.showwarning("Invalid Input", "Please enter valid numbers for hours, minutes, and seconds.")

        def show_timer_info(self, message):
            messagebox.showinfo("Timer", message)

        def set_hours(self, event=None):
            user_input = self.get_user_input("Set Hours", "Enter the number of hours:")
            if user_input is not None:
                self.hours_var.set(str(user_input).zfill(2))

        def set_minutes(self, event=None):
            user_input = self.get_user_input("Set Minutes", "Enter the number of minutes:")
            if user_input is not None:
                self.minutes_var.set(str(user_input).zfill(2))

        def set_seconds(self, event=None):
            user_input = self.get_user_input("Set Seconds", "Enter the number of seconds:")
            if user_input is not None:
                self.seconds_var.set(str(user_input).zfill(2))

        def get_user_input(self, title, prompt):
            return simpledialog.askinteger(title, prompt, parent=self.root)

        def hide_time_inputs(self):
            self.hour_label.pack_forget()
            self.hour_entry.pack_forget()
            self.minute_label.pack_forget()
            self.minute_entry.pack_forget()
            self.second_label.pack_forget()
            self.second_entry.pack_forget()

        def show_time_inputs(self):
            self.hour_label.pack(pady=5)
            self.hour_entry.pack(pady=5)
            self.minute_label.pack(pady=5)
            self.minute_entry.pack(pady=5)
            self.second_label.pack(pady=5)
            self.second_entry.pack(pady=5)

        def clear_default_value(self, var):
            if var.get() == "00":
                var.set("")

    if __name__ == "__main__":
        root = tk.Tk()
        app = TimerApp(root)
        root.mainloop()





def run_camera():
    class CameraApp:
        def __init__(self):
            # Initialize the pyttsx3 engine
            self.engine = pyttsx3.init()

            # Get a video capture object for the camera (usually default webcam)
            self.vid = cv2.VideoCapture(0)

            # Specify the directory to save captured photos
            self.photos_directory = "C:\\Users\\Asus\\OneDrive\\Pictures"

            # Specify the directory to save the recorded video
            self.videos_directory = "C:\\Users\\Asus\\OneDrive\\Pictures"

            # Create the directories if they don't exist
            os.makedirs(self.photos_directory, exist_ok=True)
            os.makedirs(self.videos_directory, exist_ok=True)

            # Variable to track if video recording is in progress
            self.recording = False

            # Variable to track if a face has been detected
            self.face_detected = False

            # Variable to track if the face detection announcement has been made
            self.face_detection_announced = False

            # Define audio parameters
            self.FORMAT = pyaudio.paInt16
            self.CHANNELS = 1
            self.RATE = 44100
            self.CHUNK = 1024

            # Load the pre-trained face detection classifier
            self.face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

            # Create a resizable window
            cv2.namedWindow('Camera Feed', cv2.WINDOW_NORMAL)
            cv2.setMouseCallback('Camera Feed', self.resize_window)

            while True:
                # Read frames from the camera
                ret, frame = self.vid.read()

                # Convert the frame to grayscale for face detection
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

                # Detect faces in the frame
                faces = self.face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

                # If a face is detected and it hasn't been detected before, announce it
                if len(faces) > 0 and not self.face_detection_announced and not self.recording:  # Add 'not recording' condition
                    self.speak_message("Face detected")
                    self.face_detection_announced = True
                elif len(faces) == 0:
                    self.face_detection_announced = False

                # Show the frames in a window
                cv2.imshow('Camera Feed', frame)

                # Add a button to capture a photo when 'c' key is pressed
                # and start/stop video recording when 'v' key is pressed
                key = cv2.waitKey(1) & 0xFF
                if key == ord('c'):
                    self.save_photo(frame)
                elif key == ord('v'):
                    if not self.recording:
                        # Start video recording
                        self.speak_message("Video recording started. Press 'v' again to stop.")
                        self.start_video_recording(frame)
                    else:
                        # Stop video recording
                        self.stop_video_recording()
                elif key == ord('q'):
                    break

                # Write the frame to the output video file when recording is in progress
                if self.recording:
                    self.output_video.write(frame)

            # Release the video capture object and close all windows
            self.vid.release()
            cv2.destroyAllWindows()

        def save_photo(self, frame):
            # Save the captured photo with a unique filename
            filename = os.path.join(self.photos_directory, f"photo_{len(os.listdir(self.photos_directory)) + 1}.jpg")
            cv2.imwrite(filename, frame)
            print(f"Photo saved as {filename}")
            # Speak out that the photo has been saved
            self.speak_message("Photo saved")

        def speak_message(self, message):
            # Speak out the provided message
            self.engine.say(message)
            self.engine.runAndWait()

        def record_audio(self, output_filename):
            audio = pyaudio.PyAudio()
            stream = audio.open(format=self.FORMAT,
                                channels=self.CHANNELS,
                                rate=self.RATE,
                                input=True,
                                frames_per_buffer=self.CHUNK)
            frames = []
            print("Recording audio...")
            while self.recording:
                data = stream.read(self.CHUNK)
                frames.append(data)
            print("Finished recording audio.")

            # Stop and close the audio stream
            stream.stop_stream()
            stream.close()
            audio.terminate()

            # Save the recorded audio to a WAV file
            wf = wave.open(output_filename, 'wb')
            wf.setnchannels(self.CHANNELS)
            wf.setsampwidth(audio.get_sample_size(self.FORMAT))
            wf.setframerate(self.RATE)
            wf.writeframes(b''.join(frames))
            wf.close()

        def start_audio_recording(self, output_filename):
            self.recording = True
            self.record_audio(output_filename)

        def stop_audio_recording(self):
            self.recording = False

        def merge_audio_video(self, video_filename, audio_filename, output_filename):
            try:
                video_clip = VideoFileClip(video_filename)
                audio_clip = AudioFileClip(audio_filename)
                video_clip = video_clip.set_audio(audio_clip)
                video_clip.write_videofile(output_filename, codec='libx264', audio_codec='aac')
                video_clip.close()
                audio_clip.close()
            except Exception as e:
                print(f"Error merging audio and video: {e}")

        def start_video_recording(self, frame):
            # Start video recording
            self.recording = True
            # Define the codec and create a VideoWriter object to save the video
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            self.output_video_filename = os.path.join(self.videos_directory, "my_gallery_video.avi")
            self.output_video = cv2.VideoWriter(self.output_video_filename, fourcc, 20.0, (640, 480))
            # Start audio recording in a separate thread
            self.audio_filename = os.path.join(self.videos_directory, "audio.wav")
            audio_thread = threading.Thread(target=self.start_audio_recording, args=(self.audio_filename,))
            audio_thread.start()
            # Update face_detection_announced to False when recording starts
            self.face_detection_announced = False

        def stop_video_recording(self):
            # Stop video recording
            self.speak_message("Video recording stopped.")
            self.output_video.release()
            self.stop_audio_recording()
            self.audio_thread.join()
            # Merge audio and video files
            merged_video_filename = os.path.join(self.videos_directory, "merged_video.mp4")
            self.merge_audio_video(self.output_video_filename, self.audio_filename, merged_video_filename)

        def resize_window(self, event, x, y, flags, param):
            if event == cv2.EVENT_MAXIMIZED:
                cv2.setWindowProperty('Camera Feed', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

    if __name__ == "__main__":
        app = CameraApp()


        
    
def main_frame_window():
    root = tk.Tk()
    root.title("Main Frame Window")

#    timer_button = tk.Button(root, text="Open Timer App", command=open_timer_app)
#    timer_button.pack(pady=10)

#    speech_to_text_button = tk.Button(root, text="Open Speech to Text App", command=open_speech_to_text_app)
#    speech_to_text_button.pack(pady=10)

    calculator_button = tk.Button(root, text="Open Calculator App", command=run_calculator)
    calculator_button.pack(pady=10)

    web_browser_button = tk.Button(root, text="Open Web Browser App", command=run_web_browser)
    web_browser_button.pack(pady=10)

    file_manager_button = tk.Button(root, text="File Manager App", command=run_file_manager)
    file_manager_button.pack(pady=10)
    
    calendar_button = tk.Button(root, text="Calendar", command=run_calendar)
    calendar_button.pack(pady=10)
    
    stopwatch_button = tk.Button(root, text="Stopwatch", command=run_stopwatch)
    stopwatch_button.pack(pady=10)
    
    speech_to_text_button = tk.Button(root, text="Text Editor App", command=run_speech_to_text_app)
    speech_to_text_button.pack(pady=10)

    timer_button = tk.Button(root, text="Timer App", command=run_timer_app)
    timer_button.pack(pady=10)
    
    camera_button = tk.Button(root, text="Camera", command=run_camera)
    camera_button.pack(pady=10)
    
    root.mainloop()



if __name__ == "__main__":
    main_frame_window()
```
